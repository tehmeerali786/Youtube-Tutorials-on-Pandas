## Indexing other than default index

```python
Transaction = pd.read_csv('data/E-Commerce.csv', encoding = "ISO-8859-1", index_col='InvoiceNo' )
```

```python
Transaction.head()
```

```python
# Indexing Operator
Transaction[0:10]
```

```python
# Indexing Operator also can be used for columns
Transaction['StockCode'].head()
```

```python
# or for list of columns
Transaction[['StockCode', 'InvoiceDate', 'CustomerID']].head()
```

```python
# However, you cannot use Indexing Operator for both rows and columns
Transaction[0:10, ['StockCode', 'InvoiceDate', 'CustomerID'] ]
```

```python
# For using rows and columns slicing together, you have to use Indexers, which are .loc or .iloc
# Since loc is used for string indexing and iloc used for number indexing, you have to convert both in numbers or strings.
Transaction.loc[0:10, ['StockCode', 'InvoiceDate', 'CustomerID']]
```

```python
Transaction.loc['536365', ['StockCode', 'InvoiceDate', 'CustomerID']]
```

```python
rows_string = Transaction.iloc[0:10].index.tolist()
rows_string
```

```python
Transaction.index
```

```python
Transaction.head(10)
```

```python
# To use index string 
Start = Transaction.index[0]
Stop = Transaction.index[9]
Start, Stop
```

```python
Transaction.loc[Start: Stop, ['StockCode', 'InvoiceDate', 'CustomerID']]
```

```python

Transaction.head()
```

```python
# Total columns of this DataFrame
Transaction.columns
```

```python
# Now you want to use numbers for both rows and columns for using indexer iloc 
col_start = Transaction.columns.get_loc('StockCode')
col_end = Transaction.columns.get_loc('Country')+1
col_start, col_end
```

```python
Transaction.iloc[0:20, col_start: col_end]
```

```python
# If you want to select particular columns
col_StockCode = Transaction.columns.get_loc('StockCode')
col_Description = Transaction.columns.get_loc('Description')
col_UnitPrice = Transaction.columns.get_loc('UnitPrice')

col_StockCode, col_Description, col_UnitPrice
```

```python
Transaction.iloc[0:20, [col_StockCode, col_Description, col_UnitPrice]]
```
